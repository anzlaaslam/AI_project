```python
from keras.models import load_model
import cv2
import numpy as np
```

```python
model = load_model('model-018.model')  # the best models that we save during training we will laod it 
 
face_clsfr=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')   # cascade classifier to get only the face part of the pic
 
source=cv2.VideoCapture(0)  ## our webcam

labels_dict={0:'MASK',1:'NO MASK'}
color_dict={0:(0,255,0),1:(0,0,255)}
```

```python
while(True):

    ret,img=source.read()   ## we will read pic frmae by frame
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  # converting image to gray as with trained data
    faces=face_clsfr.detectMultiScale(gray,1.3,5)  #3detecing faces

    for (x,y,w,h) in faces:
    
        face_img=gray[y:y+w,x:x+w]   # here we are getting only the face part of img 
        resized=cv2.resize(face_img,(100,100))  ## resizing to 100 by 100
        normalized=resized/255.0   ## normalizing it same as we did four our data training 
        reshaped=np.reshape(normalized,(1,100,100,1))
        result=model.predict(reshaped)   # passing to the model for detecting it

        label=np.argmax(result,axis=1)[0]  ## agrmax func will return coulmn for max probability (mask 0 and no mask 1)
      
        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)  # these line of code show reactangle on face with labels (mask or no mask)
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
        
    cv2.imshow('LIVE',img)
    key=cv2.waitKey(1)
    
    if(key==27):   ##esc key to break the loop
        break
        
cv2.destroyAllWindows()
source.release()
```
